```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Ex2
data = pd.read_csv('data_linear.csv')
data
```

```python
x, y = data['Diện tích'], data['Giá']
plt.scatter(x, y)
plt.xlabel('Diện tích')
plt.ylabel('Giá')
```

```python
a= np.copy(data)
a
```

```python
a[:,1] = a[:,0]
a[:,0] = 1.0
type(a)
pd.core.frame.DataFrame(a)
```

```python
def w(a,y):
  A = np.dot(np.transpose(a), a)
  b = np.dot(np.transpose(a), y)
  w = np.dot(np.linalg.pinv(A),b)
  return w
```

```python
w1 = w(a,y)
type(w1)
pd.core.frame.DataFrame(w1)
```

```python
predict = np.dot(a,w1)
predict
```

```python
plt.scatter(x, y, color='red')
plt.plot(a[:,1:],predict)
```

```python
x_predict = {'0' : [1, 1, 1], '1' : [50, 100, 150]}
x_predict = pd.DataFrame(data=x_predict)
predict = np.dot(x_predict, w1)
predict
```
